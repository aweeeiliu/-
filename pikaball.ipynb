```python
import pyautogui
import keyboard
from IPython.display import clear_output
from time import sleep, time
```

```python
# 測試各個戰鬥回合的各時間點是否有抓到這些物件
# pyautogui.locateCenterOnScreen 非常好用, 直接取中心點座標
pikachu_location = pyautogui.locateOnScreen('pikachu/pikachu.jpg', confidence = 0.2)
print(pikachu_location)
ball_location = pyautogui.locateOnScreen('pikachu/ball.jpg', confidence=0.30)
print(ball_location)
pole_location = pyautogui.locateOnScreen('pikachu/pole.jpg', confidence=0.3)
print(pole_location)
hori_kill_point = pyautogui.locateCenterOnScreen('pikachu/enemy_hori_kill.jpg', confidence=0.4)
print(hori_kill_point)
print(hori_kill_point != None)
high_kill_point = pyautogui.locateCenterOnScreen('pikachu/enemy_high_kill.jpg', confidence=0.45)
print(high_kill_point)
print(high_kill_point != None)
enemy_point = pyautogui.locateCenterOnScreen('pikachu/enemy.jpg', confidence=0.4)
print(enemy_point)
```

```python
pyautogui.PAUSE = 0.001

def left():  # 往左走,持續時間)0.3秒
    pyautogui.keyDown('left')
    sleep(0.3)
    pyautogui.keyUp('left')

    
def right(): # 往右走,持續時間0.3秒
    pyautogui.keyDown('right')
    sleep(0.3)
    pyautogui.keyUp('right')


def track(): # 追蹤各角色之間相對位置, 判斷該左該右
    
    pikachu_location = pyautogui.locateOnScreen('pikachu/pikachu.jpg', confidence=0.2)
    ball_location = pyautogui.locateOnScreen('pikachu/ball.jpg', confidence=0.3)
    pole_location = pyautogui.locateOnScreen('pikachu/pole.jpg', confidence=0.3)
    hori_kill_point = pyautogui.locateCenterOnScreen('pikachu/enemy_hori_kill.jpg', confidence=0.4)
    high_kill_point = pyautogui.locateCenterOnScreen('pikachu/enemy_high_kill.jpg', confidence=0.3)
    enemy_point = pyautogui.locateCenterOnScreen('pikachu/enemy.jpg', confidence=0.5)
    pikachu_point = pyautogui.center(pikachu_location)
    ball_point = pyautogui.center(ball_location)
    pole_point = pyautogui.center(pole_location)
    clear_output(wait=True) # 可以註解掉以追蹤各條件是否有效

    
    # 扣除掉對手撲接的圖案
    if enemy_point != None:
        # 球滯空時調整站位
        if abs(ball_point.y - enemy_point.y) >250: # 當對手還沒有要發動攻擊
            if ball_point.x < pole_point.x < pikachu_point.x: # 當球在對手那一側
                if pikachu_point.x > 1428: # 且當我的位置比我方球場中線右邊 (太靠後)
                    print('1左')
                    left() # 靠左
                elif pikachu_point.x < 1428: # 且當我的位置比我方球場中線左邊 (太靠前)
                    print('1右')
                    right() # 靠右

        # 對手殺球時預測球的落位 (not always)
        elif abs(ball_point.y - enemy_point.y) <= 200: # 當對手準備要發動攻擊
            if high_kill_point != None: # 向上擊球發生
                if pikachu_point.x < 1428: # 且我的位置比我方球場中線左邊 (太靠前)
                    print('2右')
                    right() # 靠右
                elif 1428 < pikachu_point.x : # 且我的位置比我方球場中線右邊 (太靠後)
                    print('2左')
                    left() # 靠左

            elif  hori_kill_point != None: # 平射擊球發生, 但速度好像太快來不及發生
                if pikachu_point.x < 1428: # 且我的位置比我方球場中線左邊 (太靠前)
                    print('3右')
                    right() # 靠右
                elif 1428 < pikachu_point.x : # 且我的位置比我方球場中線右邊 (太靠後)
                    print('3左')
                    left() # 靠左
            
            
if __name__ == "__main__":
    while True:
        try:
            track()
            
        except TypeError: # 當得分畫面變暗,會抓不到物件,故使用except TypeError 放他一馬
            print('oops')
            continue
        
        if keyboard.is_pressed('alt'):
            break
```
